```python
from scipy.optimize import minimize
import numpy as np
```

```python
def S(s, k, m):
    return 1 - (1 - s**k)**m
```

```python
def objective(x):
    k, m = x
    return ((S(1/3, k, m) - 1/10)**2 + (S(1/2, k, m) - 9/10)**2)
```

```python
x0 = np.array([10.0, 200.0])

bounds = [(0, 100), (0, 1000)]
```

```python
result = minimize(objective, x0, bounds=bounds)

k, m = result.x
print(f"k: {k}, m: {m}")
```

```python
print(f"S(1/2,k,m)= {(S(1/2,k,m)):0.3f} | S(1/3,k,m)= {(S(1/3,k,m)):0.3f}")
```
